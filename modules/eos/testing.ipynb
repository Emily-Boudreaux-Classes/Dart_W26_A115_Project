```python
import eos_functions as ef 
from eos_functions import * 
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm

mu = 1.004
T = np.linspace(0.0001, 1E5, 1000)
P = T
rho = ef.simple_eos(P, T, mu)

fig, ax = plt.subplots(2,2, figsize = (8,6))
ax[1,0].plot(P/np.max(P), T/np.max(T))
ax[0,0].plot(P/np.max(P),rho/np.max(rho))
ax[1,1].plot(rho/np.max(rho), T/np.max(T))
ax[1,0].set_xlabel('Pressure (normalized)')
ax[1,1].set_xlabel('Density (normalized)')
ax[1,0].set_ylabel('Temperature (normalized)')
ax[0,0].set_ylabel('Density (normalized)')
ax[0,1].set_visible(False)
plt.savefig('./figures/density-pressure-temp.png', dpi =300)
```

```python
P = np.linspace(10000, 2E16, 1000)
T = np.linspace(5000, 1.5E7, 1000)
rho = np.zeros((len(P), len(T)))
```

```python
for i in range(len(P)): 
    for j in range(len(T)): 
        rho[i,j] = ef.simple_eos(P[i], T[i], mu)
```

```python
fig,ax = plt.subplots(1,1,figsize = (8,6))
ax.contourf(T, P,np.log10(rho), levels = 5)
cbar = plt.colorbar(cm.ScalarMappable(cmap='viridis'), ax=ax)
cbar.set_label(r'$\rho$', rotation=90, fontsize = 12)
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('temperature', fontsize = 12)
ax.set_ylabel('pressure', fontsize = 12)
plt.savefig('./figures/polytropic_eos_colorplot.png', dpi = 300)
```
